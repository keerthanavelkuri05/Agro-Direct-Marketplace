```python
# Agro Direct Marketplace

products = []

while True:
    print("\n--- Agro Direct Marketplace ---")
    print("1. Add Product (Farmer)")
    print("2. View Products (Consumer)")
    print("3. Buy Product")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Enter product name: ")
        price = float(input("Enter price per kg: "))
        quantity = int(input("Enter quantity (kg): "))

        products.append({
            "name": name,
            "price": price,
            "quantity": quantity
        })

        print("Product added successfully!")

    elif choice == "2":
        if not products:
            print("No products available.")
        else:
            print("\nAvailable Products:")
            for i, p in enumerate(products, start=1):
                print(f"{i}. {p['name']} - ₹{p['price']}/kg - {p['quantity']} kg")

    elif choice == "3":
        if not products:
            print("No products available.")
        else:
            for i, p in enumerate(products, start=1):
                print(f"{i}. {p['name']} - ₹{p['price']}/kg")

            item = int(input("Select product number: ")) - 1
            qty = int(input("Enter quantity to buy (kg): "))

            if qty <= products[item]["quantity"]:
                total = qty * products[item]["price"]
                products[item]["quantity"] -= qty

                print(f"Order Successful!")
                print(f"Total Amount: ₹{total}")
            else:
                print("Insufficient stock!")

    elif choice == "4":
        print("Thank you for using Agro Direct!")
        break

    else:
        print("Invalid choice!")
```
